```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Select categorical columns
categorical_columns = training_set_values.select_dtypes(include=['object']).columns

# Compute unique counts
unique_counts = {col: training_set_values[col].nunique(dropna=False) for col in categorical_columns}

# Compute missing values count and percentage
missing_counts = training_set_values[categorical_columns].isna().sum()
missing_percent = (missing_counts / len(training_set_values)) * 100

# Create a DataFrame
unique_counts_df = pd.DataFrame({
    'Categorical Columns': categorical_columns,
    'Number of Unique Values': [unique_counts[col] for col in categorical_columns],
    'Missing Values': missing_counts.values,
    'Missing %': missing_percent.values
}).sort_values(by='Number of Unique Values', ascending=False)

# Total number of categorical columns
total_categorical_columns = len(categorical_columns)

# Plot the bar chart
plt.figure(figsize=(14, 8))
ax = sns.barplot(data=unique_counts_df, x='Number of Unique Values', y='Categorical Columns', palette="viridis")

# Add text labels on bars (HTSI applied)
for i, p in enumerate(ax.patches):
    total = int(p.get_width())  # Unique values count
    missing = unique_counts_df.iloc[i]['Missing Values']
    missing_pct = unique_counts_df.iloc[i]['Missing %']

    # Unique values count text (aligned to bar width)
    ax.annotate(f"{total}",  
                (p.get_width(), p.get_y() + p.get_height() / 2),  
                ha='left', va='center', fontsize=10, fontweight='bold', color='black')

    # Adjust horizontal shift dynamically
    shift_amount = 3500  # Fixed shift
    if total ==2697:
        shift_amount = 3200
    elif total == 356:
        shift_amount = 5000
    
    if total < 126:
        shift_amount = 5100
    # Missing values text (shifted individually)
    ax.annotate(f"(M: {missing}, {missing_pct:.1f}%)",
                (p.get_width() + shift_amount, p.get_y() + p.get_height() / 2),
                ha='left', va='center', fontsize=10, fontweight='bold', color='red')

# Labels and title
plt.xlabel("Number of Categories")
plt.xlim(0, 47000)
plt.ylabel("Categorical Columns")
plt.title(f"Number of Categories & Missing Values in Each Categorical Column\n(Total Categorical Columns: {total_categorical_columns})")

# Add legend
legend_text = "Legend:\nM: Missing Values Count\n%: Percentage of Missing Values (out of 59400)"
plt.figtext(0.49, 0.17, legend_text, fontsize=10, color="black", ha="left", bbox=dict(facecolor="orange", alpha=0.9))

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count occurrences of each category (excluding NaN by default)
category_counts = training_set_labels["status_group"].value_counts()

# Calculate the total count (excluding NaNs) and percentages
total_count = category_counts.sum()
category_percentages = (category_counts / total_count) * 100

# Count missing values and their percentage
missing_values = training_set_labels["status_group"].isna().sum()
missing_percentage = (missing_values / len(training_set_labels)) * 100

# Convert to DataFrame for plotting
category_df = category_counts.reset_index()
category_df.columns = ["Status Group", "Count"]
category_df["Percentage"] = category_percentages.values  # Add percentage column

# Plot
plt.figure(figsize=(8, 5))
ax = sns.barplot(x=category_df["Status Group"], y=category_df["Count"], palette="viridis")

# Add text labels with count and percentage on top of bars
for p, percentage in zip(ax.patches, category_df["Percentage"]):
    ax.annotate(f'{int(p.get_height())} ({percentage:.2f}%)',  # Format: Count (Percentage%)
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Add missing values info in the bottom right corner
text_info = (f"Total Rows: {len(training_set_labels)}\n"
             f"Missing Values: {missing_values} ({missing_percentage:.2f}%)")

plt.text(len(category_df["Status Group"])-0.7 , min(category_df["Count"]) * 6.5, text_info, 
         fontsize=12, verticalalignment='bottom', horizontalalignment='right',
         bbox=dict(facecolor='orange', alpha=0.8, edgecolor='black'))

# Customize the plot
plt.xlabel("Status Group")
plt.ylabel("Count")
plt.title("Distribution of Water Point Status")
plt.xticks(rotation=45)  # Rotate x labels if necessary

# Show the plot
plt.show()
```

```python
# Total number of rows in the dataset
total_rows = len(training_set_values)

# Count missing values in the "region" column
missing_values = training_set_values["region"].isna().sum()

# Calculate the percentage of missing values
missing_percentage = (missing_values / total_rows) * 100

# Exclude NaN values and count occurrences per region
region_counts = training_set_values.dropna(subset=["region"]).groupby("region").size().reset_index(name="total")

# Calculate the percentage of each region
total_count = region_counts["total"].sum()
region_counts["percentage"] = (region_counts["total"] / total_count) * 100

# Sort values in descending order
region_counts = region_counts.sort_values(by="total", ascending=False)

# Plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(y=region_counts["region"], x=region_counts["total"], color="red")

# Add text labels with total count and percentage for each bar
for p, percentage in zip(ax.patches, region_counts["percentage"]):
    ax.annotate(f'{int(p.get_width())} ({percentage:.2f}%)',  # Format: Count (Percentage%)
                (p.get_width(), p.get_y() + p.get_height() / 2), 
                ha='left', va='center', fontsize=10, fontweight='bold', color='black')

# Customize the plot
plt.xlim(0, max(region_counts["total"]) + 1500)  # Dynamically set x-axis limit
plt.xlabel("Total Count")
plt.ylabel("Region")
plt.title("Total Number of Wells by Region")

# Add total rows and missing values info in the bottom right corner
text_info = (f"Total Rows: {total_rows}\n"
             f"Missing Values: {missing_values} ({missing_percentage:.2f}%)")

# Position the text dynamically in the bottom right corner
plt.text(max(region_counts["total"]) * 1.2, 15, text_info, 
         fontsize=12, verticalalignment='bottom', horizontalalignment='right',
         bbox=dict(facecolor='orange', alpha=0.9, edgecolor='black'))

# Show the plot
plt.show()
```
