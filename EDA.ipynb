```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_status_distribution_for_top_n(merged_df, n=5):
    """
    Plots the distribution of 'status_group' for the top 'n' most common 'wpt_name' categories,
    showing both the count and percentage for each status group.

    Parameters:
    - merged_df: DataFrame containing 'wpt_name' and 'status_group'
    - n: Number of top categories to plot (default is 5)
    """
    # Find the top n 'wpt_name' categories
    top_n_wpt = merged_df['wpt_name'].value_counts().head(n).index

    # Create plots for each category
    for wpt in top_n_wpt:
        category_df = merged_df[merged_df['wpt_name'] == wpt]
        status_counts = category_df['status_group'].value_counts().reset_index()
        status_counts.columns = ['status_group', 'count']
        
        # Calculate percentage
        total_count = status_counts['count'].sum()
        status_counts['percentage'] = (status_counts['count'] / total_count) * 100

        # Plot the bar chart
        plt.figure(figsize=(10, 6))
        ax = sns.barplot(data=status_counts, x='status_group', y='count', palette="viridis")

        # Add text labels on bars (Count & Percentage)
        for p, perc in zip(ax.patches, status_counts['percentage']):
            ax.annotate(f'{int(p.get_height())}\n({perc:.1f}%)', 
                        (p.get_x() + p.get_width() / 2, p.get_height()), 
                        ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

        # Customize the plot
        plt.xlabel("Status Group")
        plt.ylabel("Count")
        plt.title(f"Distribution of Status Groups for '{wpt}'")
        plt.xticks(rotation=45)

        # Show the plot
        plt.show()

# Example usage:
plot_status_distribution_for_top_n(merged_df, n=5)  # Change n to any value you need
```
